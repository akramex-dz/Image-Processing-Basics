```python
def Func_m(image_path):
    # Import NumPy and matplotlib
    import numpy as np

    # Load an image as a NumPy array
    img = plt.imread(image_path)

    # calculer la moyenne et la déviation standard par canal
    mean_r = np.mean(img[:, :, 0]) # moyenne du canal rouge
    mean_g = np.mean(img[:, :, 1]) # moyenne du canal vert
    mean_b = np.mean(img[:, :, 2]) # moyenne du canal bleu
    std_r = np.std(img[:, :, 0]) # écart-type du canal rouge
    std_g = np.std(img[:, :, 1]) # écart-type du canal vert
    std_b = np.std(img[:, :, 2]) # écart-type du canal bleu
    
    return mean_r, mean_g, mean_b, std_r, std_g, std_b
```

```python
mean_r, mean_g, mean_b, std_r, std_g, std_b = Func_m('./figs/car.png')
print("Moyenne par canal: rouge = {:.4f}, vert = {:.4f}, bleu = {:.4f}".format(mean_r, mean_g, mean_b))
print("Déviation standard par canal: rouge = {:.4f}, vert = {:.4f}, bleu = {:.4f}".format(std_r, std_g, std_b))
```

```python
def Normalize(image_path):
    # Import NumPy and matplotlib
    import numpy as np
    import cv2
    
    # Load an image as a NumPy array
    img = plt.imread(image_path)
    
    red, green, blue = divide(image_path)
    
    redNormed = (red - red.mean())/(red.std())
    greenNormed = (green - green.mean())/(green.std())
    blueNormed = (blue - blue.mean())/(blue.std())
    
    Inorm = cv2.merge([blueNormed, greenNormed, redNormed])
    return Inorm
```

```python
Inorm = Normalize('./figs/car.png')

plt.subplot(1, 2, 1)
plt.imshow(img)
plt.title("Original Image")
plt.subplot(1, 2, 2)
plt.imshow(Inorm)
plt.title("Normalized Image")
plt.show()
```

```python
# Inverse : ayant un argument et renvoyant une variable en sortie 
# (Inv étant l’image inverse ou communément appelé le mode négatif i.e. Inv = max(I) − I ).
def Inverse(path):

    # Load an image as a NumPy array
    img = plt.imread(path)
    
    Inv = np.max(img) - img
    
    return Inv
```

```python
img = plt.imread('./figs/flower.png')
Inv = Inverse('./figs/flower.png')

plt.subplot(1, 2, 1)
plt.imshow(img)
plt.title("Original Image")
plt.subplot(1, 2, 2)
plt.imshow(Inv)
plt.title("Normalized Image")
plt.show()
```

```python
# CalcHist : ayant un argument et deux variables en sortie ( H étant un vecteur représentant l’histogramme
# de I, et b un vecteur représentant les groupes d’intervalles égaux appelés aussi bins).
def CalcHist(image_path):
    # Import NumPy and matplotlib
    import numpy as np
    from PIL import Image
    import cv2
    
    # read the image as grayscale
    img = iio.imread( uri=image_path, mode="L")

    # convert the image to float dtype with a value range from 0 to 1
    img = ski.util.img_as_float(img)
    
    H, bins = np.histogram(img, bins=256, range=(0, 1))

    return H, bins
```

```python
histogram, bin_edges = CalcHist('./figs/flower.png')

img = iio.imread( uri='./figs/flower.png', mode="L")
plt.subplot(1, 2, 1)
plt.imshow(img,  cmap="gray")
plt.title("Original Image")
plt.subplot(1, 2, 2)
plt.title("Grayscale Histogram")
plt.xlabel("grayscale value")
plt.ylabel("pixel count")
plt.xlim([0.0, 1.0])
plt.plot( bin_edges[0:-1], histogram)
plt.show()
```

```python
def FactPix(image_path, red_factor, blue_factor):
    # Read the image
    image = cv2.imread(image_path)

    # Split the image into the B, G, R channels
    B, G, R = cv2.split(image)

    # Modify the intensities of the R and B channels
    R = np.clip(R * red_factor, 0, 255).astype(np.uint8)
    B = np.clip(B * blue_factor, 0, 255).astype(np.uint8)

    # Merge the channels back into an image
    I_fact = cv2.merge([B, G, R])

    return I_fact
```

```python
I= plt.imread('./figs/car.png')
I_fact = FactPix('./figs/car.png', 0.5, 1.5)

plt.subplot(1, 2, 1)
plt.imshow(I)
plt.title('Original Image')

plt.subplot(1, 2, 2)
plt.imshow(I_fact)
plt.title('Modified Image')

plt.show()
```

```python
def Threshold(image_path, threshold):
    # Import NumPy and matplotlib
    import cv2

    # Loading the image
    img = cv2.imread(image_path)

    # Converting color mode to Grayscale
    img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
    # Thresholding the image placing 127 intensity level as threshold
    # Pixel values below 127 would be changed to Black
    # Pixel values above 127 would be changed to White (255)
    ret, T = cv2.threshold(img, threshold, 255, cv2.THRESH_BINARY)

    return T
```

```python
img = cv2.imread('./figs/flower.png')
Thresholded_img  = Threshold('./figs/flower.png', 127)

plt.subplot(1, 2, 1)
plt.imshow(img)
plt.title("Original Image")
plt.subplot(1, 2, 2)
plt.imshow(Thresholded_img, cmap='gray')
plt.title("Binary Threshold")
plt.show()
```

```python
# Une fonction définie par une série d’instructions (à partir des fonctions déjà écrites) pour pouvoir, à partir du chemin d’une image, l’ouvrir, l’afficher, afficher son histogramme, l’inverser, calculer
# son histogramme après inversion et visualiser le résultat des traitements
def func_j(image_path):

    #read the original image 
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

    #plot the original image
    plt.figure(figsize=(12, 6))
    plt.subplot(3, 2, 1)
    plt.imshow(image, cmap='gray')
    plt.title('Original Image')
    
    #histogram of the original image
    H, bins = CalcHist(image_path)
    plt.subplot(3, 2, 2)
    plt.xlabel("grayscale value")
    plt.ylabel("pixel count")
    plt.xlim([0.0, 1.0])
    plt.plot( bins[0:-1], H)

    #the inverse of the image
    Inverted_image_new = Inverse(image_path)
    plt.subplot(2, 2, 3)
    plt.imshow(Inverted_image_new, cmap='gray')
    plt.title('Inverted Image')

    # Saving the inverted image
    inverted_image_path = './figs/inverted_image_new.png'
    cv2.imwrite(inverted_image_path, Inverted_image_new)

    #histogram of the inversed image
    H_inv, bins_inv = CalcHist(inverted_image_path)
    plt.subplot(2, 2, 4)
    plt.xlabel("grayscale value")
    plt.ylabel("pixel count")
    plt.xlim([0.0, 1.0])
    plt.plot( bins_inv[0:-1], H_inv)

    plt.tight_layout()
    plt.show()
```

```python
func_j('./figs/flower.png')
```

```python
def Func_t(image_path):
    
    # Open and display the image
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    plt.figure(figsize=(12, 6))
    plt.subplot(2, 2, 1)
    plt.imshow(image, cmap='gray')
    plt.title('Original Image')

    # Calculate and display the histogram of the image
    H, bins = CalcHist(image_path)
    plt.subplot(2, 2, 2)
    plt.xlabel("grayscale value")
    plt.ylabel("pixel count")
    plt.xlim([0.0, 1.0])
    plt.plot( bins[0:-1], H)
    plt.title('Histogram of Original Image')

    # Normalize the image to the range [10, 50]
    normalized_image = Normalize(image_path)

    #saving the normalized image
    #because CalcHist function takes the path of the image as a parameter, we need to save the normalized image to a file and then pass the path of the file to the CalcHist function
    normalized_image_path = './figs/normalized_image.png'
    cv2.imwrite(normalized_image_path, normalized_image)

    # Calculate and display the histogram of the normalized image
    H_norm, bins_norm = CalcHist(normalized_image_path)
    plt.subplot(2, 2, 3)
    plt.imshow(normalized_image, cmap='gray')
    plt.title('Normalized Image')

    # Display the histogram of the normalized image
    plt.subplot(2, 2, 4)
    plt.xlabel("grayscale value")
    plt.ylabel("pixel count")
    plt.xlim([0.0, 1.0])
    plt.plot( bins_norm[0:-1], H_norm)
    plt.title('Histogram of Normalized Image')

    plt.tight_layout()
    plt.show()
```

```python
Func_t('./figs/flower.png')
```

```python
def Func_f(image_path):

    # Open and display the image
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    plt.figure(figsize=(12, 6))
    plt.subplot(2, 2, 1)
    plt.imshow(image, cmap='gray')
    plt.title('Original Image')

    # Calculate and display the histogram of the image
    H, bins = CalcHist(image_path)
    plt.subplot(2, 2, 2)
    plt.xlabel("grayscale value")
    plt.ylabel("pixel count")
    plt.xlim([0.0, 1.0])
    plt.plot( bins[0:-1], H)
    plt.title('Histogram of Original Image')

    # Threshold the image with a threshold value of 128
    thresholded_image = Threshold(image_path, 128)

    #saving the thresholded image
    thresholded_image_path = './figs/thresholded_image.png'
    cv2.imwrite(thresholded_image_path, thresholded_image)

    # Calculate and display the histogram of the thresholded image
    H_thresh, bins_thresh = CalcHist(thresholded_image_path)
    plt.subplot(2, 2, 3)
    plt.imshow(thresholded_image, cmap='gray')
    plt.title('Thresholded Image')

    # Display the histogram of the thresholded image
    plt.subplot(2, 2, 4)
    plt.xlabel("grayscale value")
    plt.ylabel("pixel count")
    plt.xlim([0.0, 1.0])
    plt.plot( bins_thresh[0:-1], H_thresh)
    plt.title('Histogram of Thresholded Image')

    plt.tight_layout()
    plt.show()    
```
